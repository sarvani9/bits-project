```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE

# Load the data
cosm = pd.read_csv("datasets/cosmetic_preprocessed.csv")

# Check the first five rows 
display(cosm.sample(5))

# Inspect the types of products
display(cosm['Label'].value_counts())
```

```python
# Split the ingredients with '\r\n\r\n'
a = [t.split('\r\n\r\n') for t in cosm['ingredients']]
```

```python
# Take onle the part with the real ingredients 
pattern = ['\r\n', '-\w+: ', 'Please', 'No Info', 
           'This product','Visit']
for i in range(len(cosm)):
    num = len(a[i])
    for j in range(num):
        if all(x not in a[i][j] for x in pattern):
           cosm['ingredients'][i] = a[i][j]
```
